```python
# remove words within parenthesis
# remove symbols
no_symbol = mother

import re
regex = re.compile(".*?\((.*?)\)")
within_parenthesis = re.findall(regex, no_symbol)

for i in within_parenthesis:
    no_symbol = no_symbol.replace(i, '')
    
no_symbol = no_symbol.replace('(', '')
no_symbol = no_symbol.replace(')', '')
no_symbol = no_symbol.replace(',', '')
no_symbol = no_symbol.replace('/', ' ')
```

```python
no_symbol
```

```python
# Remove "Part of Sentences"
word = no_symbol

parts = ['indefinite article', 'definite article', 'infinitive marker', 'modal v.', 'auxiliary v.', 'prep.', 'adv.', 'pron.', 'det.', 'adj.', 'conj.', 'prep', 'number', 'exclam.', 'n.', 'v.']

for part in parts:
    word = word.replace(part, '')
    
word = word.replace('    ', ' ')
word = word.replace('   ', ' ')
word = word.replace('  ', ' ')
```

```python
# Split string of words into list
word_list = word.split(' ')
word_list = [s for s in word_list if s != '']
```

```python
# Making a list of part of sentences from no_symbol
temp = no_symbol
PoS = []

for x in range(0,len(word_list)):
    if x != len(word_list) - 1:
        temp = temp.replace(word_list[x], '', 1)
        PoS.append(temp.split(word_list[x+1])[0])
        temp = temp.replace(temp.split(word_list[x+1])[0], '', 1)
    else:
        temp = temp.replace(word_list[x], '', 1)
        PoS.append(temp + ' ')
        break
```

```python
# Replacing abbreviations into full words
for i in range(0, len(PoS)):
    PoS[i] = PoS[i].replace(' ', '', 1)
    PoS[i] = PoS[i].replace('prep.', 'preposition')
    PoS[i] = PoS[i].replace('adv.', 'adverb')
    PoS[i] = PoS[i].replace('n.', 'noun')
    PoS[i] = PoS[i].replace('v.', 'verb')
    PoS[i] = PoS[i].replace('adv.', 'adverb')
    PoS[i] = PoS[i].replace('modal v.', 'modal verb')
    PoS[i] = PoS[i].replace('auxiliary v.', 'auxiliary verb')
    PoS[i] = PoS[i].replace('pron.', 'pronoun')
    PoS[i] = PoS[i].replace('det.', 'determiner')
    PoS[i] = PoS[i].replace('adj.', 'adjective')
    PoS[i] = PoS[i].replace('conj.', 'conjunction')
    PoS[i] = PoS[i].replace('exclam.', 'exclamation')
    PoS[i] = PoS[i][:-1]
    PoS[i] = PoS[i].replace(' ', '; ')
    PoS[i] = PoS[i].replace('infinitive; marker', 'infinitive marker')
    PoS[i] = PoS[i].replace('indefinite; article', 'indefinite article')
    PoS[i] = PoS[i].replace('definite; article', 'definite article')
    PoS[i] = PoS[i].replace('auxiliary; verb', 'auxiliary verb')
    PoS[i] = PoS[i].replace('modal; verb', 'modal verb')
    
PoS
```

```python
# Turning it into DF to export it as an excel file
result = pd.DataFrame(list(zip(word_list, PoS)), 
               columns =['Words', 'Part of Sentence'])
```
